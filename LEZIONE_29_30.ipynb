```python
#INDICIZZARE UNA SPECIFICA COLONNA <DATE>: Ho voluto riscrivere la funzione perchè volevo esercitarmi, ma avrei potuto semplicemente inserire il modulo "index" nella funzione già sviluppata sopra (come nel video-corso)
def prepocessing (name):
  df =pd.read_csv (name, delimiter = "\t", index_col= "<DATE>", parse_dates= True)
  #ATTENZIONE: Andare a rivedere l'utilizzo della funzione "parse_dates"
  df = df.iloc [:,:-2]
  return df
df = prepocessing ("/EURUSD_D1.csv")
df
```

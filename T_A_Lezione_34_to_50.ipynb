```python
#LEZIONE_36: VECTORIZED BAEDCKTEST STRATEGY
#Adesso andiamo a vedere come backtestare tramite il VECTORIZED BACKTEST per misurare
#la validitÃ  della nostra strategia e la sua performance

import numpy as np

#Creiamo una colonna dove andremo ad inserire i risultati: "RETURNS"
df["returns"]=np.log(df.close.div(df.close.shift(1)))
df["strategy"]=df.shift(1) * df["returns"]
df

#Cancelliamo nuovamente i dati NaN (ovvero mancanti)
df.dropna(inplace=True)
```

```python
#Absolute Performance 1

df[["returns","strategy"]].sum()

#Absolut Performance 2
df[["returns","strategy"]].sum().apply(np.exp)

#Annualized Return
df[["returns","strategy"]].mean()*252

#Annualized Risk
df[["returns","strategy"]].std()*np.sqrt(252)

#Cumulatives returns
df["creturns"]=df["returns"].cumsum().apply(np.exp)
df["strategy"]=df["strategy"].cumsum().apply(np.exp)
```

IMPORTANTE: Il codice che andremo a vedere adesso ci permette di testare una nostra strategia (in questo esempio la strategia SMA delle lezioni precedenti) potendo semplicemente cambiare le variabili dell'SMA per valutare , considerando un periodo definito, i risultati e quindi quali siano le variabili migliori per quel periodo

```python
#LEZIONE_37: Finding the optimal SMA Strategy

def run_strategy (SMA):
  df=df.copy()
  df["returns"]=np.log(df.close.div(df.close.shift(1)))
  df["SMA_S"]=df.close.rolling(int(SMA[0])).mean()
  df["SMA_L"]=df.close.rolling(int(SMA[1])).mean()
  df.dropna(inplace= True)

  df["position"]=np.where(df["SMA_S"]>df["SMA_L"],1,-1)
  df["strategy"]=df.position.shift(1) * df["returns"]

  return df[["returns","strategy"]].sum().apply(np.exp)[-1]

  run_strategy ([50,200])

#Adesso importiamo da Scypy la funzione BRUTE. Questo ci permette di testare
#un numero di combinazioni variabile con un'unico codice

  from scypy.optimize import brute

brute(run_strategy,((10,50),(20,60)))
```

LEZIONE_43:OOP-The PLOT_RESULTS Method

```python
class SMABacktester():
  def __init__(self,symbol,SMA_S,SMA_L,start,end):
    self.symbol=symbol
    self.SMA_S=SMA_S
    self.SMA_L=SMA_L
    self.start=start
    self.end=end
    self.results= None
    self.data = None
    #self.get_data = None

  def get_data(self):
    raw=pd.read_csv("/content/forex_pairs.csv",parse_dates=["Date"], index_col="Date")
    raw = raw[self.symbol].to_frame().dropna()  # Fix: Removed extra raw
    raw = raw.loc["2004-01-01":self.end].copy()  # Fix: Use self.start and self.end
    raw = raw.rename(columns={self.symbol:"price"})  # Fix: Removed inplace=True
    raw["return"]=np.log(raw["price"]/raw["price"].shift(1))
    raw["SMA_S"]=raw["price"].rolling(window=self.SMA_S).mean()
    raw["SMA_L"]=raw["price"].rolling(window=self.SMA_L).mean()
    self.data=raw
    return raw

  def set_parameters(self,SMA_S=None,SMA_L=None):
    if SMA_S is not None:
      self.SMA_S=SMA_S
      self.data["SMA_S"]=self.data["price"].rolling(window=self.SMA_S).mean()
    if SMA_L is not None:
      self.SMA_L=SMA_L
      self.data["SMA_L"]=self.data["price"].rolling(window=self.SMA_L).mean()

  def test_strategy(self):
    data=self.data.copy().dropna()
    data["position"]=np.where(data["SMA_S"]>data["SMA_L"],1,-1)
    data["strategy"]=data["position"].shift(1) * data["return"]
    data.dropna(inplace= True)
    data["creturns"]=data["return"].cumsum().apply(np.exp) # Fix: Use data["return"] instead of data["returns"]
    data["cstrategy"]=data["strategy"].cumsum().apply(np.exp)
    perf = data["cstrategy"].iloc[-1] #absolute performance
    outperf = perf-data["creturns"].iloc[-1] #outperformance
    self.results = data[["creturns", "cstrategy"]] # Store results for plotting
    return round(perf,6),round(outperf,6)

     # Define plot_results method within the class
  def plot_results(self):
    if self.results is None:
      print ("Nessun risultato da plottare adesso. Ripeti il test")
    else:
      title="{} | SMA_S={} | SMA_L={}".format(self.symbol,self.SMA_S,self.SMA_L)
      self.results[["creturns","cstrategy"]].plot(title=title,figsize=(12,8))

  def update_and_run(self,SMA):
    self.set_parameters(int(SMA[0]),int(SMA[1]))
    return self.test_strategy()[0]

  def optimize_parameters (self, SMA_S_range, SMA_L_range):
    opt=brute(self.update_and_run,((SMA_S_range[0],SMA_S_range[1]),(SMA_L_range[0],SMA_L_range[1])),finish=None)
    return opt
```

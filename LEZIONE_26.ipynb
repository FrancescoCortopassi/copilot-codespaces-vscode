```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#simulazione di dati
arr= np.random.randint (0,15, size= (15,))

#square of array
square= arr **2

#Multiplication of array
mut = arr*3

print (arr)
print (square)
print(mut)
```

ABBIAMO CREATO DEI DATI: Per plottare dei grafici dobbiamo ovviamente avere dei dati su cui lavorare

```python
#plottare la data
plt.plot (arr)
```

```python
plt.plot (arr)
plt.plot (square)
plt.plot (mut)
#abbiamo prvato a plottare 3 differenti dati sullo stesso grafico (arr, square, mut)
```

```python
#cambiare la label X
plt.xlabel ("Random Numbers")
```

```python
#cambiare la label Y
plt.ylabel ("Esmeralda Gate")
```

```python
#Cambiare il titolo
plt.title ("Le follie dell' Imperatore")
```

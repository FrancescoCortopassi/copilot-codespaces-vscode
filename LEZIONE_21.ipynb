**MODULO RANDOM per ARRAY:** Questa parte è molto importante in finanza perchè molot spesso le cose in finanza seguono un processo randomizzato

```python
import numpy as np
random_array = np.random.rand (6)
random_array
```

Come possiamo vedere Python ha creato una serie di numeri random in un array

```python
ran = np.random.rand (5)
print (f"Shape is {np.shape(ran)}")

ran =ran.reshape(-1,1)
print (f"Shape is {np.shape(ran)}")
```

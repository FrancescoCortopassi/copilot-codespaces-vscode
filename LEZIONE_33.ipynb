```python
#SMA: In questa lezione andiamo a creare il primo TRADING SYSTEM basato su una MEDIA MOBILE SEMPLICE (SMA)

!pip install yfinance
#prima di tutto importiamo le librerie necessarie

import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings ("ignore")

```

```python
df= yf.download ("EURUSD=X")
df
```

```python
df.columns= ["open","high","low","close","volume"]
df
```

```python
#SMA: Creare una Simple Moving Average a 30 giorni
#per creare la SMA FAST e la SLOW SMA si utilizzano delle variabili, si definisce su quale colonna deve agire la "finestra"
#e poi si usa il modulo ROLLING
df["SMA fast"]= df ["close"].rolling(30).mean ()
df["SMA slow"]= df ["close"].rolling(60).mean ()
```

```python
#PLOTTARE il risultato
df[["close","SMA fast","SMA slow"]].loc ["2020"].plot (figsize =(15,8))
```

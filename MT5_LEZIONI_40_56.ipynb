LEZIONE_45: BACKTEST KPI
In questa lezione andiamo a vedere quali sono i KPI importanti per un backtest di base eseguito in maniera professionale. nelle prossime lezioni vedremo poi come implementare questi KPI all'interno del codice.


1. ABOSLUTE RETURN (pip)
The total amount of return for a trading strategy. We calculate that in PIPS (for forex market)

2. WIN RATE (percentage)
Number of operations with a positive outcome/operations with negative outcome.
Better described: number of win positions/all positions

3. MEAN RETUNR PER TRADE
It's the mean of the winnings trade sum in a specifica period. It could be also calculated for losing trades with the same structure

4. MAIMUM CONSECUTIVES GAINS/LOSES
This metric mesure the long streack (striscia) of wins/loses in a period

5. MAXIMUM DRAWDOWN
Largest drop in asset price over a specific timeperiod in the equity of an account (the difference between a peack and a trough in the line curve of an asset)

LEZIONE_53: TRIPLE RIS STRATEGY INTRO
Questa strategia usa un'indicatore molto conosciuto da tutti itrader tecnici e molto apprezzato anche da me: l'RSI.
L'acronimo di RSI sta per Relative Strenhgt Index, e misura la forza relativa dei movimenti di un'asset segnalando i livelli di OVERSOLD ed OVERBUY.
Questo tipo di indicatore è definito oscillatore perché è sviluppato per lavorare all'interno di un range. La parte alta di questo range indica l'oversold e viversa. Questo indicatore è funzionale per indicare eventuali cambi di trend o l'indebolimento del trend presente, ed è efficace soprattutto con strategie di MEAN REVERSION come quella che andremo a vedere adesso.

Generalmente il range dell'RSI è settato tra 30 (range inferiore) e 70 (range superiore), e la strategia che andremoa d analizzare ed implementare sarà settata su questo range.

ENTRY LOGIC:
.The 5-day RSI is below 30
.The 5-day RSI reading is down for the third day in a row
.The 5-day RSI reading was below since 60 three trading days ago
.The close is higher than the 200-day moving average

If all this conditions are met, a buy signal is generated
A sell signal is generated when the 5-days RSI cross over 50

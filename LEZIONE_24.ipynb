```python
#Importazione del file CSV
import pandas as pd
assets = pd.read_csv('/content/assets (1).csv')
assets
#ho importato il file cvs trascinandolo dalla cartella dei DOWNLOAD al folder di GITHUB (pratica da considerare STANDARD)
```

```python
#Parzializzazione dei dati: non ho compreso bene l'operatività....DA RIVEDERE
assets = pd.read_csv('/content/assets (1).csv',parse_dates=True, index_col= "time")
assets
```

```python
#adesso invertiamo l'ordine della cronologia temporale (che Python ci da di default dalla data più recente alla più vecchia)
assets= assets.sort_index(ascending= True)
assets
```

```python
#Selezionare una colonna
assets= assets [["Close DJI30","Close CAC40","Close SP500"]]
assets
```

```python

#Funzione DROPNA (è la funzione specifica per la cancellazine dei valori mancanti, documentarsi su pandas per i suoi vari utilizzi)
assets= assets.dropna ()
assets
```

```python
#Resettare l'indice
assets = assets.reset_index (drop = True)
assets
```

```python
#Rolling
#questo procedimento è interessante peer la cre4azione di strumenti di ANALISI TECNICA (da approfondire in seguito)
assets ["SMA 15"] = assets ["Close DJI30"].rolling(15).mean()
assets
```

```python
assets = assets.dropna ()
assets
```

```python
#Shift
#questo metodo è molto interessante quando si vogliono creare la percentuale della variazione (PER IL CALCOLO DEL RITORNO DI UNA STRATEGIA)
assets ["Close SP500"].shift(1)
assets
```

ATTENZIONE: Non ho compreso in maniera chiara questa funzione SHIFT, molto importante per definire il RITORNO DI UNA STRATEGIA: dvovrà essere rivista e compresa nel primo ripasso utile

```python
#Funzione GROUPBY.In questo caso viene utilizzata per calcolare la media di una determinata colonna, direi molto interessante in applicazione finanziaria
df= pd. DataFrame (assets, columns =["Close SP500"])
df.groupby (by= "Close SP500"). mean()
```

```python
#Groupby by sum
df.groupby (by= "Close SP500"). sum()
```

```python
#Groupby by std (standard deviation)
df.groupby (by= "Close SP500"). std()
```

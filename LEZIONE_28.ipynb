```python
#Ricreiamo nuovamente l'ambiente di lavoro
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
#simulazione di dati
arr= np.random.randint (0,15, size= (15,))

#square of array
square= arr **2

#Multiplication of array
mut = arr*3

print (arr)
print (square)
print(mut)
```

```python
plt.scatter (arr,mut)
plt.scatter (arr, square)
```

```python
#Adattare la dimensione del grafico
plt.figure (figsize= (15,8))
```

```python
#Plottare la linea
plt.plot (arr, alpha= 0.6, color = "turquoise")
```

```python
#Provo semplicemente a cambiare colore
plt.plot (arr, alpha= 0.6, color = "yellow")
```

```python
plt.plot (square, "-0",linewidth=3 )
```

```python
plt.figure (figsize= (15,8))
plt.plot (arr, alpha= 0.6, color = "turquoise")
plt.plot (square, "-o",linewidth=3, color = "green" )
plt.plot (mut, "o", color= "orange" )
plt.legend (["Turquoise", "Square", "Mut"])
```

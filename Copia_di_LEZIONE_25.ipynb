```python
import pandas as pd
assets = pd.read_csv('/content/assets (1).csv')
assets
```

```python
#Slicing
assets.iloc [0: 1500, 0: 10]
```

```python
assets.iloc [0: 1500, 0: 4]
#come possiamo notare il primo valore che assegniamo è quello delle ROWS, il secondo è quello delle COLUMNS). Questo, com egià visto in precedenza, è uno STANDARD in Python
```

```python
#Slicing condizionato da un valore
assets.loc [assets["Close DJI30"]> 1500]
```

```python
#Condizionare dalla data
assets.loc ["2010":"2012"]
#NOTARE BENE: Tra 2 valori che ho passato come vaqriabili per il LOC ho inserito i ":" e non la "," e questo perché io ho interrogato per avere un'intervallo, e non solo per segnalare due valori
```

IMPORTANTE: Devo comprendere meglio come utilizzare le varie parentesi (tonde, quadre o graffe) per passare i valori. Una volta compreso questo concetto avrò compreso una parte importantissima della sintassi in Pithon

```python

#Concatenare by the column
df1= assets.iloc [:, : 3]
df2= assets.iloc [:,3:]

pd.concat ((df1,df2), axis=1)
```

```python
#Concatenare by the row
df1= assets.iloc [:, : 1500]
df2= assets.iloc [:,1500:]

pd.concat ((df1,df2), axis=1)
```
